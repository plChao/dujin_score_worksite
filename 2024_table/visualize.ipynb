```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

article_df = pd.read_csv('./article_info.csv')
matrix = {}
for col in ['A', 'B', 'C', 'D', 'E', 'F', 'G']:
    matrix[col] = {}
for index, row in article_df.iterrows():
    matrix[row['article_id'][0]][row['article_id'][1:]] = 0
```

```python
actual_df = pd.read_csv('./actual_exam_situation.csv')
for index, row in actual_df.iterrows():
    matrix[row['article_id'][0]][row['article_id'][1:]] += 1
print(matrix)
```

```python
article_counts = actual_df.groupby('article_id').size().reset_index(name='article_count')
article_cnt_name = article_counts.merge(article_df, how='right')
article_cnt_name.sort_values(by='article_count', ascending=False, inplace=True)
article_cnt_name.to_csv('./vis_art_name.csv', index=False)
```
